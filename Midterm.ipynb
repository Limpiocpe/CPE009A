<h1><b>#24</h1></b>
Write a function that checks if a string of parentheses (), brackets [], and braces {} is balanced.

```python
def check_bs(s):
    if len(s) % 2 != 0:
        return False

    pair = {')': '(', ']': '[', '}': '{'}
    stack = []

    for c in s:
        if c in '([{':
            stack.append(c)
        elif c in ')]}':
            if not stack or stack[-1] != pair[c]:
              return False
            stack.pop()
    return len(stack) == 0

print(check_bs("({[]})"))
print(check_bs("([)]"))
print(check_bs("((()))"))
print(check_bs("({)}"))
print(check_bs(input("Enter brackets: ")))
```
