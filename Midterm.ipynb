```python
x = 1
for y in range(5):
  for y in range(x):
    print("*", end="")
  print()
  x+=1

for y in range(4):
  for y in range(4-y):
    print("*", end="")
  print()
```

```python
i = 0

check = str(input())
print(len(check))
for x in range (len(check)):
  j = len(check)
  y = i + 1
  z = j - 1
  if check[i:y] == check[z:j]:
    i+=1
    j-=1
    if i == len(check):
      print("True")
  else:
    print("False")
```
