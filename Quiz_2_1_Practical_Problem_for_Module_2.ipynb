```python
# We want to create a python code for a grading scheme

# Create a function to accept a number of activities
# Compute for the score vs total
# Return the average grade of the student in %
def termInput(act_no, activity_type):
  rawScore = []
  totalScore = []

  for i in range(act_no):
    while True:
      try:
        input_score = float(input(f"{activity_type} #{i+1} Score: "))
        input_total = float(input(f"{activity_type} #{i+1} Total: "))
        print()

        if input_score > input_total:
          print("\nInvalid Scores: Score cannot be greater than the total. Please re-enter.\n")

        elif input_total == 0:
          print("\nInvalid Total: Total Score cannot be zero. Please re-enter.\n")

        else:
          rawScore.append(input_score)
          totalScore.append(input_total)
          break
      except ValueError:
        print("Invalid input. Please enter a valid number.")

  if not totalScore or sum(totalScore) == 0:
    return 0.0
  else:
    grade = sum(rawScore) / sum(totalScore)
    grade = grade * 100
    return grade
```

```python
# Create a function for the class standing per term
def termCS(term_name):
  print(f"Enter the number of Activities for {term_name} in the following: ")
  termHOA = termInput(int(input(f"{term_name} Hands-on Activities:")), "Hands-on Activity")

  termQuiz = termInput(int(input(f"{term_name} Quizzes: ")), "Quiz")

  termAss = termInput(int(input(f"{term_name} Assignments: ")), "Assignment")

  grade = (0.5 * termHOA) + (0.3 * termQuiz) + (0.2 * termAss)
  return grade
```

```python
# Create a function to compute the term grade
def termGrade(term_name):
  print(f"Computing for the {term_name} term grade.")
  examGrade = 0.0
  while True:
    try:
      examRaw = float(input(f"{term_name} Exam Score: "))
      examTotal = float(input(f"{term_name} Exam Total: "))
      print()

      if examTotal == 0:
        print(f"\n{term_name} Exam Total cannot be zero. Please re-enter the total again.\n")
      else:
        examGrade = (examRaw/examTotal) * 100
        break
    except ValueError:
      print("Invalid input. Please enter a valid number.")

  CSgrade = termCS(term_name)

  grade = (0.5 * examGrade) + (0.5 * CSgrade)
  return grade
```

```python
# Create a function to compute for the grades in the semester
def semesterGrade():
  print("Prelim Grade")
  prelimGrade = termGrade("Prelim")

  print("Midterm Grade")
  midtermGrade = termGrade("Midterm")
  finalMG = (1/3)*prelimGrade + (2/3)*midtermGrade

  print("Final Grade")
  finalGrade = termGrade("Final")
  finalFG = (1/3)*finalMG + (2/3)*finalGrade

  return finalFG
```

```python
def convertGrade(finalGrade):
  print("The Final Grade is ")
  if finalGrade >= 99.0: return 1.0
  #The class should add everything else
  elif finalGrade >= 96.0: return 1.25
  elif finalGrade >= 93.0: return 1.5
  elif finalGrade >= 90.0: return 1.75
  elif finalGrade >= 87.0: return 2.0
  elif finalGrade >= 84.0: return 2.25
  elif finalGrade >= 81.0: return 2.5
  elif finalGrade >= 78.0: return 2.75
  elif finalGrade >= 75.0: return 3.0
  elif finalGrade < 75: return "Failed"
```
