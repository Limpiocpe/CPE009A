# **LOOP Statement**

#Example 1 (For Loop statement)

```python
week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
for x in week:
  print(x)
```

#Example 2 (The break statement)

```python
week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
for x in week:
  print(x)
  if x == 'Thursday':
    break
```

#Example 3 (The break statement)
#+1 Point



```python
week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
for x in week:
  if x == 'Thursday':
    break
  print(x)
```

#Example 4 (Looping through a string)

```python
for x in "week":
  print(x)
```

The range() function

```python
for x in range(6):
  print(x)
print()
for x in range(2,6):
  print(x)
print()
```

```python
for y in range(2, 6, 2):
  print(y)
```

Nested Loops

```python
adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adjective:
  for y in fruits:
    print(x, y)
```

While Loops

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

The break statement

```python
i = 1
while i < 6:
  print(i)
  if i ==3:
    break
  i += 1
```

Continue Statement

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

#+1 Point Midterm

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    print(i)
  continue
```

The else Statement

```python
i = 1
while i < 6:
  print(i)
  i +=1
else:
  print("i is no longer less than 6")
```
