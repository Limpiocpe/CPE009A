<h2>Distance Formula</h2>
1. Make a program that would calculate the distance between two points given a list of coordinates. Use the distance  formula.
coorindates_list = [(1,1), (2,3)]


```python
import math

coordinates_list = [(1,1), (2,3)]

x1, y1 = coordinates_list[0]
x2, y2 = coordinates_list[1]

distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print(f"The coordinates are: {coordinates_list}")
print(f"The distance between the two points is: {distance:.2f}")
```

<h2>Simple Word Filter</h2>
2. For a given string input, replace all the words "stupid" with an asterisk * equal to the length of the string. The new  string value should be displayed with the asterisks.

```python
input_string = input("Enter a string: ")
word_to_replace = "stupid"
replacement_asterisks = '*' * len(word_to_replace)

filtered_string = input_string.replace(word_to_replace, replacement_asterisks)

print(f"Original string: {input_string}")
print(f"Filtered string: {filtered_string}")
```

<h2>Phonebook</h2>
3. Create a simple phonebook program that can read from a list of dictionaries. The program should be able to display a  person"s name, contact, and address based from a user input which is the id of the record.

```python
phonebook_entries = [
    {"id": 1, "name": "Jeon Heejin", "contact": "09123456789", "address": "Block 1 Lot 12 ARTMS Subdivision"},
    {"id": 2, "name": "Kim Seoyeon", "contact": "09921223131", "address": "Modhaus Dorm Room 105"},
    {"id": 3, "name": "Yves", "contact": "09324238223", "address": "12 Apple Street"}
]

def find_person_by_id(person_id, entries):
    for person in entries:
        if person["id"] == person_id:
            return person
    return None

while True:
    try:
        search_id = int(input("Enter the ID of the person you want to find (or 0 to exit): "))
        if search_id == 0:
            print("Exiting phonebook.")
            break

        found_person = find_person_by_id(search_id, phonebook_entries)

        if found_person:
            print(f"\nName: {found_person['name']}")
            print(f"Contact: {found_person['contact']}")
            print(f"Address: {found_person['address']}")
        else:
            print(f"Person with ID {search_id} not found.")
    except ValueError:
        print("Invalid input. Please enter a number.")
```
