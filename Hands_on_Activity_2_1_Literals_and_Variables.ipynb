**Procedure:**

A teacher wants to calculate the final grade in a CpE course and want to write it in a python program. The following are the requirements:
1. PRELIM GRADE = 50% Prelim Exam + 50% Prelim Class Standing (CS)
2. PRELIM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
3. MIDTERM GRADE = 1/3 of PRELIM GRADE + 2/3 of (50% Midterm Exam + 50% Midterm Class Standing (CS))
4. MIDTERM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
5. FINAL GRADE = 1/3 of MIDTERM GRADE + 2/3 of (50% Final Exam + 50% Final Class Standing (CS))
6. FINAL CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
7. HOAs, Quizzes and Assignments are inputted as average of all submissions and are out of 100%.
8. Major exams are inputted out of 100%.
9. Show the codes that successfully run the program.
10. Provide comments or documentation strings for your program.

```python
def calculate1():
  # Place all the output of the procedures here
  #Accepts input and Computes Prelim Class Standing
  p_hoa = float(input("Enter Prelim Hands-on Activity Grade: "))
  p_quiz = float(input("Enter Prelim Quiz Grade: "))
  p_as = float(input("Enter Prelim Assignment Grade: "))
  p_cs = (p_hoa*.5 + p_quiz*.3 + p_as*.2)
  print("Prelim CS is: ", "%.2f" % p_cs)
  print()

  p_exam = float(input("Enter Prelim Exam Score: "))
  p_grade = (p_exam*.5 + p_cs*.5)
  print("Prelim Grade is: ", "%.2f" % p_grade)
  print()

  #Accepts input for Midterm Class Standing Computation
  m_hoa = float(input("Enter Midterm Hands-on Activity Grade: "))
  m_quiz = float(input("Enter Midterm Quiz Grade: "))
  m_as = float(input("Enter Midterm Assignment Grade: "))
  m_cs = (m_hoa*.5 + m_quiz*.3 + m_as*.2)
  print("Midterm CS is: ", "%.2f" % m_cs,)
  print()

  m_exam = float(input("Enter Midterm Exam Score: "))
  #Computes the Midterm grade by getting the 1/3 of the Prelim Grade and 2/3 of Midterm Exam and Class Standing
  m_grade = ((m_exam*.5 + m_cs*.5)*(2/3) + (p_grade * (1/3)))
  print("Midterm Grade is: ", "%.2f" % m_grade)
  print()

  #Accepts input for Final Class Standing Computation
  f_hoa = float(input("Enter Final Hands-on Activity Grade: "))
  f_quiz = float(input("Enter Final Quiz Grade: "))
  f_as = float(input("Enter Final Assignment Grade: "))
  f_cs = (f_hoa*.5 + f_quiz*.3 + f_as*.2)
  print("Final CS is: ", f_cs)
  print()

  f_exam = float(input("Enter Final Exam Score: "))
  f_grade = ((f_exam*.5 + f_cs*.5)*(2/3) + (m_grade * (1/3)))
  print("Final Grade is: ", "%.2f" % f_grade)
  print()
calculate1()
```

```python
# Place all the output of the supplementary activity here
students = []
for i in range(3):
    name = input(f"Enter name of student {i+1}: ")
    calculate1()
    students.append((name, p_grade, m_grade, f_grade))
    print()

print("Student Grades:")
for name, p_grade, m_grade, f_grade in students:
    print(f"Name: {name}")
    print(f"Prelim Grade: {p_grade:.2f}")
    print(f"Midterm Grade: {m_grade:.2f}")
    if f_grade >= 99 < 100:
      print("Final Grade is 1.00")
    elif f_grade >= 96 <= 99:
      print("Final Grade is 1.25")
    elif f_grade >= 93 <= 95:
      print("Final Grade is 1.50")
    elif f_grade >= 90 <= 92:
      print("Final Grade is 1.75")
    elif f_grade >= 87 <= 89:
      print("Final Grade is 2.00")
    elif f_grade >= 84 <= 86:
      print("Final Grade is 2.25")
    elif f_grade >= 81 <= 83:
      print("Final Grade is 2.50")
    elif f_grade >= 78 <= 80:
      print("Final Grade is 2.75")
    elif f_grade >= 75 <= 77:
      print("Final Grade is 3.00")
    else :
        print("Final Grade is 5.00")
```
