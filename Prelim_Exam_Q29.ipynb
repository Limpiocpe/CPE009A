Develop a python scripts that:

Calculate total marks and percentage
Assign grade for each subject based on marks:

90-100: A+

80-89: A

70-79: B+

60-69: B

50-59: C

40-49: D

Below 40: F (Fail)

Determine overall result (Pass/Fail)


Performance Remarks:


Percentage ≥ 90: "Outstanding"

Percentage ≥ 80: "Excellent"

Percentage ≥ 70: "Very Good"

Percentage ≥ 60: "Good"

Percentage ≥ 40: "Satisfactory"

Below 40: "Needs Improvement"

```python
def calculate_marks(subject_marks_list, subjects):
    total_marks = sum(subject_marks_list)
    percentage = (total_marks / (subjects * 100)) * 100

    individual_subject_grades = []
    for mark in subject_marks_list:
        if 90 <= mark <= 100:
            individual_subject_grades.append("A+")
        elif 80 <= mark < 90:
            individual_subject_grades.append("A")
        elif 70 <= mark < 80:
            individual_subject_grades.append("B+")
        elif 60 <= mark < 70:
            individual_subject_grades.append("B")
        elif 50 <= mark < 60:
            individual_subject_grades.append("C")
        elif 40 <= mark < 50:
            individual_subject_grades.append("D")
        else:
            individual_subject_grades.append("F (Fail)")

    overall_result = "Pass"
    for mark in subject_marks_list:
        if mark < 40:
            overall_result = "Fail"
            break

    performance_remarks = ""
    if percentage >= 90:
      performance_remarks = "Outstanding"
    elif percentage >= 80:
      performance_remarks = "Excellent"
    elif percentage >= 70:
      performance_remarks = "Very Good"
    elif percentage >= 60:
      performance_remarks = "Good"
    elif percentage >= 40:
      performance_remarks = "Satisfactory"
    else:
      performance_remarks = "Needs Improvement"

    return total_marks, percentage, individual_subject_grades, overall_result, performance_remarks

def main():
  subjects = int(input("Enter the number of subjects: "))
  subject_input = []
  subject_name = []

  for i in range(subjects):
    while True:
      try:
        name = input(f"Enter the name of Subject #{i+1}: ")
        subject_name.append(name)
        grade = float(input(f"Enter the marks for {subject_name[i]} (out of 100): "))
        if 0 <= grade <= 100:
              subject_input.append(grade)
              break
        else:
            print("Invalid input. Please enter a number between 0 and 100.")
      except ValueError:
          print("Invalid input. Please enter a number.")

  total_marks, percentage, grade, overall_result, performance_remarks = calculate_marks(subject_input, subjects)
  t_mark = (100 * subjects)

  print(f"\nStudent Report")
  for i in range(subjects):
    print(f"{subject_name[i]}: {subject_input[i]}")
    print(f"Grade: {grade[i]}")
  print("------------------------")
  print(f"Total Marks: {total_marks:.0f}/{t_mark}")
  print(f"Percentage: {percentage:.2f}%")
  print(f"Overall Result: {overall_result}")
  print(f"Performance Remarks: {performance_remarks}")
```

```python
print("=============================")
print("Trial #1 (For Passing)")
main()
print("\n=============================")
print("Trial #2 (For Failed)")
main()
```
